Random Forest

```r
data(iris)
print("Structure of Dataset: Iris")
str(iris)
print("Installation and loading of caTools and randomForest")
install.packages("caTools")
install.packages("randomForest")
library(caTools)
library(randomForest)
print("Splittinf the dataset into train and test samples")
split<-sample.split(iris,SplitRatio=0.7)
train<-subset(iris,split=="TRUE")
test<-subset(iris,split=="FALSE")
print("Setting the seed so that the random values calculation starts from 120")
set.seed(120)
print("Building the randomForest model")
classif<-randomForest(x=train[-5],y=train$Species,ntree=500)
print("Pridicting the values for test sample")
ypred<-predict(classif,newdata=test[-5])
print("Building the confussion matrix")
confi<-table(test[,5],ypred)
confi
print("Plotting the values of the classifier built")
plot(classif)
print("Plotting the importance")
importance(classif)
```

Clustering Kmeans

```r
install.packages("dplyr")
install.packages("ggplot2")
install.packages("ggfortify")
library("dplyr")
library("ggplot2")
library("ggfortify")
summary(iris)
head(iris)
data<-select(iris,c(1:4))
# We need only the datapoints without the output variables for clustering as the output will be decidid by the ethod

wssplot<- function(data,nc=15,seed=1234) {
  wss<-(nrow(data)-1)*sum(apply(data,2,var)) # apply() is used to apply var(varriance) function on data, 2 means column based 1 means row based
  for (i in 2:nc) {
    set.seed(seed)
    wss[i]<-sum(kmeans(data,centers=i)$withinss)
  }
  plot(1:nc,wss,type="b",xlab="no of clusters",ylab="within group sum squares")
  wss
}
wssplot(data)
kmean<-kmeans(data,3)
kmean$centers
autoplot(kmean,data,frame=TRUE)
```

Logistic Regression

```r
# install.packages("dplyr")
# install.packages("caTools")
install.packages("ROCR")
# library("dplyr")
# library("caTolls")
library("ROCR")
summary(mtcars)
split<-sample.split(mtcars,SplitRatio=0.7)
train<-subset(mtcars,split==TRUE)
test<-subset(mtcars,split==FALSE)
log.m<-glm(vs~gear+disp,data=train,family="binomial")
summary(log.m)
predreg<-predict(log.m,newdata=test,type="response")
predreg<-ifelse(predreg>0.5,1,0)
table(test$vs,predreg)
miss<-mean(predreg!=test$vs)
print(paste('Accuracy',1-miss))
ROCPred<-prediction(predreg,test$vs)
ROCPer<-performance(ROCPred,measure="tpr",x.measure="fpr")
plot(ROCPer)
auc<-performance(ROCPred,measure='auc')
auc<-auc@y.values[[1]]
auc
plot(ROCPer,customize=TRUE,
print.cutoffs.at=seq(0.1,by=0.1),
main="ROC Curve")
abline(a=0,b=1)

auc<-round(auc,4)
legend(.6,.4,auc,title="AUC",cex=1)
```
